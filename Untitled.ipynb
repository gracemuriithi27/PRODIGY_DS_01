```python
import pandas as pd

file_path = 'API_SP.POP.TOTL_DS2_en_csv_v2_900.csv'
data = pd.read_csv(file_path, skiprows=4)

missing_data_summary = data.isna().sum()
missing_percentages = (missing_data_summary / len(data)) * 100

years_columns = [col for col in data.columns if col.isdigit()]
missing_data_years = missing_percentages[years_columns]

print("Missing Data Summary for Years:")
print(missing_data_years.sort_values(ascending=False))
```

```python
from sklearn.linear_model import LinearRegression
import numpy as np

def fill_missing_with_regression(data, target_year):

    years = [str(year) for year in range(1960, 2024)]
    available_data = data.dropna(subset=years)  
    target_data = data[data[target_year].isna()] 

    if target_data.empty:
        print(f"No missing data for the year {target_year}.")
        return data

    X_train = available_data[years].drop(columns=[target_year])
    y_train = available_data[target_year]
    X_predict = target_data[years].drop(columns=[target_year])

    model = LinearRegression()
    model.fit(X_train, y_train)

    target_data[target_year] = model.predict(X_predict)
    filled_data = pd.concat([available_data, target_data])
    
    return filled_data
```

```python
import pandas as pd
from sklearn.linear_model import LinearRegression
from sklearn.pipeline import Pipeline
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import StandardScaler
from sklearn.impute import SimpleImputer

def fill_missing_with_regression(data, target_year):
    print(f"Missing values in {target_year} before filling: {data[target_year].isna().sum()}")
    
    missing_data = data[data[target_year].isna()]
    
    if missing_data.empty:
        print(f"No missing data in {target_year}.")
    else:
        data_dropped = data.dropna(subset=[target_year])

        # Prepare X (features) and y (target)
        X = data_dropped.drop([target_year, 'Country Name', 'Country Code', 'Indicator Name', 'Indicator Code'], axis=1)
        y = data_dropped[target_year]

        preprocessor = ColumnTransformer(
            transformers=[
                ('num', Pipeline([
                    ('imputer', SimpleImputer(strategy='mean')),  
                    ('scaler', StandardScaler())  
                ]), X.columns)
            ]
        )
        
        pipeline = Pipeline(steps=[
            ('preprocessor', preprocessor),
            ('regressor', LinearRegression())
        ])
        
        pipeline.fit(X, y)
        
        X_missing = missing_data.drop([target_year, 'Country Name', 'Country Code', 'Indicator Name', 'Indicator Code'], axis=1)
        
        predicted_values = pipeline.predict(X_missing)
        
        data.loc[data[target_year].isna(), target_year] = predicted_values
        print(f"Missing values in {target_year} after filling: {data[target_year].isna().sum()}")
    
    return data

target_year = '2021'
data_filled = fill_missing_with_regression(data, target_year)

print(f"\nMissing values in {target_year} after filling: {data_filled[target_year].isna().sum()}")
```

```python
import pandas as pd

df = pd.read_csv('Linear Regression cleaned_data.csv') 

df_long = df.melt(id_vars=['Country Name', 'Country Code', 'Indicator Name', 'Indicator Code'],
                  var_name='Year', value_name='Population')

df_long['Year'] = pd.to_numeric(df_long['Year'], errors='coerce')
```

```python
df_long['Population'] = df_long['Population'].fillna(df_long['Population'].median())
```
